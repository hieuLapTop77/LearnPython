```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Loaddata = pd.read_csv('cereal.csv')
Loaddata.head()
```

```python
fields = ['shelf','weight','cups','rating']
data = Loaddata.drop(fields,axis=1)
data.head()
```

```python
data_corr = data.corr()
data_corr
```

```python
# sử dụng ones_like để build matrix thành dạng bool (T,F)
ones_corr = np.ones_like(data_corr, dtype=bool)
ones_corr
```

```python
# sử dụng hàm triu trong numpy để lấy một nửa giá trị bên dưới
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data = data_corr);
```
